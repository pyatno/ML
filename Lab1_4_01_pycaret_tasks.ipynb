Изучить как использовать PyCaret для решения задачи регрессии

**1. Установка PyCaret**

```python
pip install --pre pycaret
```

**2. Загрузка данных**

```python
# Импорт нужных функций
from pycaret.datasets import get_data
from pycaret.regression import setup, compare_models, evaluate_model, predict_model

data = get_data('insurance')
data.head()  # Посмотрим первые строки
```

**3. Настройка эксперимента**

```python
from pycaret.regression import setup

exp = setup(
    data=data,          # Данные
    target='charges',   # Целевая переменная
    session_id=123      # Для воспроизводимости
)
```

**4. Сравнение моделей**

```python
best_model = compare_models()
```

**5. Оценка модели**

```python
evaluate_model(best_model)
```

**6. Предсказания**

```python

predictions = predict_model(best_model, data=data)
predictions[['charges', 'prediction_label']].head()
```
